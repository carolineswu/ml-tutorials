### Entropy and gain

You might be wondering: how are decision trees constructed? How does the model decide on what is the most ideal split during fitting? To answer this we'll need to use a value called **entropy**. If we have $N$ classes, and $p_i$ is the probability of randomly selecting an example in class $i$: $$entropy = -\sum_{i=1}^{N} p_i log_2 p_i$$

Entropy is a way to measure the "surprise" of a random process. For instance, the outcome of a fair coin is very uncertain, so it has high entropy. We can reasonably guess the outcome of a biased coin (ex: heads with probability 0.9, tails 0.1), so it has low entropy. Below is a plot of entropy when there are only two classes. As we expect, entropy is maximized at p=0.5 and minimized at the extremes.</br></br>
![Entropy plot](entropy_plot.png)

We'll also need to use the idea of **information gain**:
$$Gain = E_{parent} - E_{children}$$
where $E_{parent}$ is the entropy of the parent node, and $E_{children}$ is the average entropy of the child nodes.

We want to decrease the entropy as much as possible from the parent to the children, since this means we're reducing the uncertainty of our choice as much as possible - we definitely want this when constructing a model! So the decision tree chooses the feature split that maximizes the gain.

In practice, we don't actually have to use entropy when building decision trees - lots of concave functions work fine. We often want to substitute another function since calculating the log term in entropy is time-consuming. The default function for sklearn decision trees is 'gini', which is faster than entropy and almost as accurate:

$$gini = 1-\sum_{i=1}^{N} {p_i}^2 $$

What's wrong with our model?

![Bias-variance tradeoff](bias_variance_tradeoff.png)

Having a training accuracy that is way higher than test accuracy is a classic sign of overfitting. Recall that high bias means we're underfitting (e.g. fitting a linear model to quadratic data), and high variance means we're overfitting. The figure below demonstrates how a super complicated decision tree results in a complicated decision boundary that isn't representative of reality.

![Overfitting decision boundary](overfitting_decision_boundary.png)

Decision trees by themselves, especially if they're allowed to grow unchecked, have low bias but high variance and are at risk of overfitting.

One way we limit tree size is by setting max_depth. Another way is enforcing there must be a least a certain number of points in a leaf node - this hyperparameter is called `min_samples_leaf` in sklearn (usually leaf nodes must be "pure", meaning all points belong to the same class. When we set `min_samples_leaf` and allow leaf nodes with points from multiple classes, we take a majority vote to classify points.) 

We call models like decision trees "weak learners." We can combine a bunch of weak learners to get a stronger learner - this is the general idea behind a class of algorithms called **ensemble learning**. "Sometimes the average opinion of a bunch of idiots is better than the opinion of one expert." This is where random forests come in!

### Random forests

Random forests take the average output of many decision trees, which lowers variance (reduces overfitting) while maintaining low bias. We also don't want our decision trees to be too similar, otherwise our variance won't decrease as much. So we have to introduce randomness into our decision trees somehow. Random forests achieve this in two ways: 
1. Train decision trees on random samples of training data
2. At each treenode, take a random sample of m features (out of d total features). We are only allowed to split on one of the m features
    - m is a hyperparameter we can choose! For classification, $m=\sqrt{d}$ works well.

There's a lot of hyperparameters involved in decision trees. Check out the [`RandomForestClassifier` documentation](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html) for default values. Here are the hyperparameters you should definitely know:
- n_estimators (number of trees in the forest)
- max_depth
- min_samples_leaf
- max_features (the m value from before)
- max_samples (size of the random sample)
- bootstrap (default True, meaning we sample with replacement)
